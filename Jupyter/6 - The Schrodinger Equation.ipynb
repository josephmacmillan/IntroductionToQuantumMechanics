## 6.3 Superposition States

Basic time dependence in Example 6.1

```python
%matplotlib inline

import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np

rc('text.latex',preamble='\\usepackage{libertine}\n\\usepackage[libertine]{newtxmath}')
rc('font',**{'family':'serif','serif':['Linux Libertine O']}, size=18)
rc('text', usetex=True)
```

```python
t = np.linspace(0, 6*np.pi, 1000)
P = 0.5 * (1 + np.cos(t))

fig = plt.figure(figsize=(8, 4))
ax = fig.add_subplot(1, 1, 1)

ax.plot(t, P, color='black')
ax.set_xlabel(r'$\omega t$')
ax.set_ylabel(r'$\mathcal{P}_{+a}$')

plt.savefig('fig_simple_time_evolution.pdf', bbox_inches='tight')
```
