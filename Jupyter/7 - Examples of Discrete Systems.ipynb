# 7 - Examples of Discrete Systems
## 7.1 Electron in a Magnetic Field

Time dependent probability for measuring $S_x = \hbar/2$.  I'll take $\theta = \pi/6$.

```python
%matplotlib inline

import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np

rc('text.latex',preamble='\\usepackage{libertine}\n\\usepackage[libertine]{newtxmath}')
rc('font',**{'family':'serif','serif':['Linux Libertine O']}, size=18)
rc('text', usetex=True)
```

```python
t = np.linspace(0, 6*np.pi, 1000)
theta = np.pi/6
P = 0.5 * (1 + np.sin(theta) * np.cos(t))

fig = plt.figure(figsize=(8, 4))
ax = fig.add_subplot(1, 1, 1)

ax.plot(t, P, color='black')
ax.set_xlabel(r'$\omega_0 t$')
ax.set_ylabel(r'$\mathcal{P}_{+x}$')
ax.set_xlim(0, 20)
ax.set_ylim(0, 1)

ax.hlines([0.5*(1-np.sin(theta)), 0.5 * (1+np.sin(theta))], 0, 20, linestyle='--', color='red')

plt.savefig('fig_simple_time_evolution.pdf', bbox_inches='tight')
```
