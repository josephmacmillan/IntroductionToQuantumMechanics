```python
%matplotlib inline

import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np
from scipy import special

rc('text.latex',preamble='\\usepackage{libertinus}')
rc('font',**{'family':'serif','serif':['Linux Libertine O']}, size=18)
rc('text', usetex=True)
```

```python
def draw_wave(ax):

    z = special.ai_zeros(100)[0]

    x = np.linspace(0, 30, 1000)
    y = 4*special.airy(x + z[20])[0]

    def rotate(x, y, angle):
        xp = x * np.cos(angle) - y * np.sin(angle)
        yp = x * np.sin(angle) + y * np.cos(angle)

        return xp, yp

    ax.plot(x, y, lw=2, color='#d90429ff')
    for a in np.arange(-0.01, 0.01, 0.0025):
        xp, yp = rotate(x, y, a)
        ax.plot(xp, yp, lw=1, color='#d90429ff')

    
fig = plt.figure(figsize=(6, 9))
ax = fig.add_axes([0, 0, 1, 1])

z = special.ai_zeros(100)[0]

x = np.linspace(0, 30, 1000)
y = 4*special.airy(x + z[20])[0]
ax.fill_between(x, y, y2 = -40, color='#8d99aeff')
ax.fill_between(x, y, y2 = 30, color='#edf2f4ff')

draw_wave(ax)

ax.set_ylim(-35, 25)
ax.set_xlim(0, 30)
ax.axis('off')

#ax.text(1, 5, "Introduction to Quantum Mechanics", fontsize=30)
#ax.text(4.5, -5, "Joseph D. MacMillan", fontsize=20)

plt.savefig('fig_cover1.pdf')
```

```python
  
fig = plt.figure(figsize=(6, 0.5))
ax = fig.add_axes([0, 0, 1, 1])

draw_wave(ax)
ax.axis('off')

plt.savefig('fig_cover_wave.pdf')
```
