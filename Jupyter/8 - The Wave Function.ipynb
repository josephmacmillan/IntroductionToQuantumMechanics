# 8 - The Wave Function
## 8.2 Position Indeterminancy

An example wave funciton:

```python
%matplotlib inline

import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np

rc('text.latex',preamble='\\usepackage{libertine}\n\\usepackage[libertine]{newtxmath}')
rc('font',**{'family':'serif','serif':['Linux Libertine O']}, size=18)
rc('text', usetex=True)
```

```python
x = np.linspace(-2, 5, 1000)
psi = np.abs(np.exp(-np.abs(1.2*x))*np.cos(x)*(-0.01*x**3-0.1*x**2-0.2*x+0.1))**2

fig = plt.figure(figsize=(8, 4))
ax = fig.add_subplot(1, 1, 1)

ax.plot(x, psi, color='black')
ax.text(5.1, -0.001, r'$x$')
ax.text(0.2, 0.012, r'$|\psi(x)|^2$')

ax.spines[['left', 'bottom']].set_position('zero')
ax.spines[['top', 'right']].set_visible(False)

ax.set_xticks([])
ax.set_yticks([])

ax.text(-0.4, -0.0015, "A")
ax.text(1.5, -0.0015, "B")
ax.text(3.2, -0.0015, "C")

ax.set_xlim(-2, 5)
ax.set_ylim(-0.002, 0.014)

plt.savefig('fig_example_wave_function.pdf', bbox_inches='tight')
```

Collapsed wave function:

```python
x = np.linspace(-2, 5, 1000)
psi = np.abs(0.12*np.exp(-1000*(x-3.3)**2))**2

fig = plt.figure(figsize=(8, 4))
ax = fig.add_subplot(1, 1, 1)

ax.plot(x, psi, color='black')
ax.text(5.1, -0.001, r'$x$')
ax.text(0.2, 0.012, r'$|\psi(x)|^2$')

ax.spines[['left', 'bottom']].set_position('zero')
ax.spines[['top', 'right']].set_visible(False)

ax.set_xticks([])
ax.set_yticks([])

ax.text(-0.4, -0.0015, "A")
ax.text(1.5, -0.0015, "B")
ax.text(3.2, -0.0015, "C")

ax.set_xlim(-2, 5)
ax.set_ylim(-0.002, 0.014)

plt.savefig('fig_collapsed_wave_function.pdf', bbox_inches='tight')
```
